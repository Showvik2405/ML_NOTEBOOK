Pandas provides two types of classes for handling data:

***Series:*** a one-dimensional labeled array holding data of any type 
such as integers, strings, Python objects etc.
***DataFrame:*** a two-dimensional data structure that holds data like
a two-dimension array or a table with rows and columns.

```python
import pandas as pd
import numpy as np

# Create sample DataFrame
df = pd.DataFrame({
    "A": [1, -2, 3, -4, 5, -7, 9],
    "B": [10, 20, 30, 40, 50, 60, 70],
    "C": ["w", "z", "y", "x", "v", "p", "q"]
}, index=["a", "c", "b", "d", "e", "f", "g"])

# Display the entire DataFrame
print("DataFrame:\n", df)

# 1. First 5 rows (default)
print("\n1. df.head():\n", df.head())

# 2. Last 2 rows
print("\n2. df.tail(2):\n", df.tail(2))

# 3. Row index
print("\n3. df.index:\n", df.index)

# 4. Column names
print("\n4. df.columns:\n", df.columns)

# 5. Convert to NumPy array
print("\n5. df.to_numpy():\n", df.to_numpy())

# 6. Summary statistics (only numeric columns)
print("\n6. df.describe():\n", df.describe())

# 7. Transpose DataFrame
print("\n7. df.T:\n", df.T)

# 8. Sort columns in descending alphabetical order
print("\n8. df.sort_index(axis=1, ascending=False):\n", df.sort_index(axis=1,ascending=False))

# 9. Sort rows by values in column 'B'
print("\n9. df.sort_values(by='B'):\n", df.sort_values(by="A"))

# 10. Select column 'A'
print("\n10. df['A']:\n", df["A"])

# 11. Select first 3 rows (by position)
print("\n11. df[0:3]:\n", df[0:3])

# 12. Select columns 'A' and 'B' using label-based indexing
print("\n12. df.loc[:, ['A', 'B']]:\n", df.loc[:, ["A", "B"]])

# 13. Filter rows where column 'A' > 0
print("\n13. df[df['A'] > 0]:\n", df[df["A"] > 0])
```

```python
import pandas as pd
import numpy as np

# Create a date range
dates = pd.date_range("2025-01-01", periods=6)

# Create a sample DataFrame
df = pd.DataFrame(np.random.randn(6, 4), index=dates, columns=list("ABCD"))

print("Original DataFrame:\n", df)

# Reindex to select only the first 4 dates and add a new column "E"
df1 = df.reindex(index=dates[0:4], columns=list(df.columns) + ["E"])

print("\nReindexed DataFrame (df1):\n", df1)

# Assign value 1 to column 'E' for the first two dates
df1.loc[dates[0]:dates[1], "E"] = 1

print("\nDataFrame after assigning 1 to 'E' column for first two dates:\n", df1)
```

```python
#Handling missing value

# Create a DataFrame with missing values
data = {
    'A': [1, 2, np.nan, 4],
    'B': [np.nan, 2, 3, 4],
    'C': [1, np.nan, np.nan, 4]
}
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# 1. Detect missing values
print("\n1. Check for missing values (True = missing):\n", df.isna())

# 2. Count missing values in each column
print("\n2. Count of missing values per column:\n", df.isnull().sum())

# 3. Drop rows with any missing values
print("\n3. Drop rows with any missing values:\n", df.dropna(how = 'any'))

# 4. Drop columns with all missing values
print("\n4. Drop columns where all values are missing:\n", df.dropna(axis=1, how='all'))

# 5. Fill missing values with a constant (e.g., 0)
print("\n5. Fill missing values with 0:\n", df.fillna(0))

# 6. Fill missing values with column mean
print("\n6. Fill missing values with column mean:\n", df.fillna(df.mean(numeric_only=True)))

# 7. Forward fill (propagate previous value down)
print("\n7. Forward fill:\n", df.fillna(method='ffill'))

# 8. Backward fill (propagate next value up)
print("\n8. Backward fill:\n", df.fillna(method='bfill'))
```

```python
#Statistical Operations
import pandas as pd
import numpy as np

# Sample data
data = {
    'A': [10, 20, 30, 40, 50],
    'B': [7, 14, 21, 28, 35],
    'C': [100, np.nan, 300, 400, 500]
}
df = pd.DataFrame(data)
print("DataFrame:\n", df)

# Summary statistics
print("\n1. Summary statistics:\n", df.describe())

# Mean
print("\n2. Mean:\n", df.mean(numeric_only=True))

# Median
print("\n3. Median:\n", df.median(numeric_only=True))

# Mode
print("\n4. Mode:\n", df.mode(numeric_only=True))

# Standard Deviation
print("\n5. Standard Deviation:\n", df.std(numeric_only=True))

# Variance
print("\n6. Variance:\n", df.var(numeric_only=True))

# Min and Max
print("\n7. Min:\n", df.min(numeric_only=True))
print("\n8. Max:\n", df.max(numeric_only=True))
```

```python
#Filtering
import pandas as pd
import numpy as np

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 35, 40, 28],
    'Score': [85, 62, 90, 70, 95]
}
df = pd.DataFrame(data)

# Add a missing value for demonstration
df.loc[2, 'Score'] = np.nan

print("Original DataFrame:\n", df)

# 1. Filter rows where Age > 30
print("\n1. Age > 30:\n", df[df['Age'] > 30])

# 2. Filter rows where Score is between 70 and 90
print("\n2. Score between 70 and 90:\n", df[(df['Score'] >= 70) & (df['Score'] <= 90)])

# 3. Filter rows where Name is 'Alice' or 'Eva'
print("\n3. Name is 'Alice' or 'Eva':\n", df[df['Name'].isin(['Alice', 'Eva'])])

# 4. Filter rows where Name is NOT 'Bob'
print("\n4. Name is NOT 'Bob':\n", df[df['Name'] != 'Bob'])

# 5. Filter rows where Score is NaN (missing)
print("\n5. Score is NaN:\n", df[df['Score'].isna()])

# 6. Filter rows where Score is NOT NaN
print("\n6. Score is NOT NaN:\n", df[df['Score'].notna()])

# 7. Filter using query method (Age < 35 and Score > 80)
print("\n7. Query: Age < 35 and Score > 80:\n", df.query("Age < 35 and Score > 80"))
```

```python
#sorting
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['A', 'B', 'C', 'D', 'E'],
    'Age': [25, 30, 35, 40, 28],
    'Score': [85, 62, 90, 70, 95]
}
df = pd.DataFrame(data)

print("Original DataFrame:\n", df)

# 1. Sort by a single column (Score) ascending
print("\n1. Sort by Score (ascending):\n", df.sort_values(by='Score'))

# 2. Sort by a single column (Score) descending
print("\n2. Sort by Score (descending):\n", df.sort_values(by='Score', ascending=False))

# 3. Sort by multiple columns: first by Age, then by Score
print("\n3. Sort by Age, then Score:\n", df.sort_values(by=['Age', 'Score']))

# 4. Sort by index (default ascending)
print("\n4. Sort by index (row labels):\n", df.sort_index())

# 5. Sort columns (column names) in descending order
print("\n5. Sort columns (descending):\n", df.sort_index(axis=1, ascending=False))
```

```python
import pandas as pd

# Sample DataFrame
data = {
    'Department': ['Sales', 'Sales', 'HR', 'HR', 'IT', 'IT'],
    'Employee': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Salary': [50000, 55000, 60000, 62000, 70000, 72000],
    'Bonus': [5000, 6000, 4000, 4500, 7000, 7500]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# 1. Group by single column and calculate mean
print("\n1. Average Salary per Department:\n", df.groupby('Department')['Salary'].mean())

# 2. Group by multiple columns
print("\n2. Total Salary and Bonus per Department:\n", df.groupby('Department')[['Salary', 'Bonus']].sum())

# 3. Get multiple statistics using .agg()
print("\n3. Multiple aggregations (mean and max) on Salary:\n", df.groupby('Department')['Salary'].agg(['mean', 'max']))

# 4. Count number of employees in each department
print("\n4. Employee count per Department:\n", df.groupby('Department')['Employee'].count())

# 5. Group and get entire rows of max salary per department
print("\n5. Highest paid employee in each Department:\n",
      df.loc[df.groupby('Department')['Salary'].idxmax()])

# 6. Iterate through groups
print("\n6. Iterating through groups:")
for dept, group in df.groupby('Department'):
    print(f"\nDepartment: {dept}")
    print(group)
```
