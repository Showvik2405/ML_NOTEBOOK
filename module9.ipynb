# catagorical scatterplot , stripplot

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', jitter=False, data=tips) # jitter=False means no random noise in y axis
```

```python
sns.stripplot(x='day', y='total_bill', jitter=False, data=tips, hue= 'sex')
```

```python
sns.stripplot(x='day', y='total_bill', jitter=False, data=tips, palette='coolwarm') # coolwarm is a color palette
```

# Swarmplot solve gittering problem

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex') # Swarmplot solve gittering problem
```
